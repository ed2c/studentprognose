```python
data = pd.read_excel("//cnas.ru.nl/wrkgrp/TeamIR/Man_info/Student Analytics/Prognosemodel RU/Syntax/Python/studentprognose/data/input/totaal_old.xlsx")
```

```python
def read_json(file_path="../configuration/configuration.json"):
    f = open(file_path)

    data = json.load(f)

    f.close()

    return data

configuration = read_json()
numerus_fixus_list = list(configuration["numerus_fixus"].keys())

NFs = configuration["numerus_fixus"]
for year in data["Collegejaar"].unique():
    for week in data["Weeknummer"].unique():
        for nf in NFs:
            nf_data = data[(data["Collegejaar"] == year) & (data["Weeknummer"] == week) & (data["Croho groepeernaam"] == nf)]

            if np.sum(nf_data['Prognose_ratio']) > NFs[nf]:
                nf_data[nf_data["Herkomst"] == "NL"]["Prognose_ratio"] = nf_data[nf_data["Herkomst"] == "NL"]["Prognose_ratio"] - (
                    np.sum(nf_data['Prognose_ratio']) - NFs[nf]
                )

            if np.sum(nf_data['SARIMA_individual']) > NFs[nf]:
                nf_data[nf_data["Herkomst"] == "NL"]["SARIMA_individual"] = nf_data[nf_data["Herkomst"] == "NL"]["SARIMA_individual"] - (
                    np.sum(nf_data['SARIMA_individual']) - NFs[nf]
                )

            if np.sum(nf_data['SARIMA_cumulative']) > NFs[nf]:
                nf_data[nf_data["Herkomst"] == "NL"]["SARIMA_cumulative"] = nf_data[nf_data["Herkomst"] == "NL"]["SARIMA_cumulative"] - (
                    np.sum(nf_data['SARIMA_cumulative']) - NFs[nf]
                )

print(data.columns)
```

```python
# Code to replace data_latest with forecasted data

data = pd.read_excel("../data/output/output_prelim.xlsx")
data_latest = pd.read_excel("//cnas.ru.nl/wrkgrp/TeamIR/Man_info/Student Analytics/Prognosemodel RU/Syntax/Python/studentprognose/data/input/totaal.xlsx")

def replace_latest_data(data_latest: pd.DataFrame, data: pd.DataFrame):
    years_used = data["Collegejaar"].unique()
    weeks_used = data["Weeknummer"].unique()

    for year in years_used:
        for week in weeks_used:
            data_latest = data_latest[~((data_latest["Collegejaar"] == year) & (data_latest["Weeknummer"] == week))]

    data_latest = pd.concat([data_latest, data], ignore_index=True)

    return data_latest

data_latest = replace_latest_data(data_latest, data)
data_latest.to_excel("output_prefinal.xlsx", index=False)
```
